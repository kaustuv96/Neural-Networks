### Business Goal:

Build a predictive model to identify undecided voters.

```python
import pandas as pd
raw_df=pd.read_excel(r"D:\Business Analytics\Important files\Neural networks\DemoRepublic.xlsx")
```

```python
#-------------------------------------------------------------------
##  Select the Predictor Variables into a dataframe called X_features
##--------------------------------------------------------------------
X_features = list( raw_df.columns )
X_features.remove( 'Undecided' )
```

```python
print(X_features)
```

```python
##------------------------------------------
##  Set up a dataframe with only X-variables
##------------------------------------------
X =  raw_df[X_features]
```

```python
##------------------------------------------
##  Identify the Class Label or Y variable
##------------------------------------------
Y = raw_df.Undecided.astype('category')

print(Y.info())
```

```python
##-----------------------------------------
##  Split the Data into Training and Test
##  test size=0.3 indicates 70% training and
##  30% test data
##-----------------------------------------
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
Y,
test_size = 0.3,
random_state = 500)
print(y_train.info())

##############################################################################
##  All data processing should be done after split to prevent data leakage  ##
##############################################################################
```

```python
##############################################################################
##  Pre-process X-vars   ##
##############################################################################

##-----------------------------------------------------------------------
##  1. Identify and dummy code (one-hot encoding) the categorical column
##-----------------------------------------------------------------------
# List of categorical columns
categorical_columns = ['HomeOwner','Female','Married','Church']
X_train_encoded = pd.get_dummies(X_train, columns=categorical_columns,drop_first = True,dtype=int )
X_test_encoded = pd.get_dummies(X_test, columns=categorical_columns,drop_first = True ,dtype=int)
```

```python
#The purpose of pre-processing in this context is to transform the raw categorical data into a format that machine learning algorithms can interpret and use effectively. 
# 1. Transform categorical variables into numeric format (0/1).
#2. Avoid multicollinearity by dropping redundant categories.
# 3. Ensure consistency between training and testing datasets.
#4. Improve model interpretability and predictive performance.
# 5. Increase model performance.
# In raw form, categorical variables are often non-numeric, which machine learning models (e.g., linear regression, decision trees, neural networks) cannot interpret directly. One-hot encoding ensures that the model treats each category distinctly and appropriately, which is critical for building effective models.
```

```python
##-----------------------------------------------------------------------
##  2. Normalise the cont. columns
##-----------------------------------------------------------------------
from sklearn.preprocessing import MinMaxScaler
continuous_columns= ['Age','Income','Education','HouseholdSize']
scaler_min_max = MinMaxScaler()
X_train_encoded[continuous_columns] = scaler_min_max.fit_transform(X_train_encoded[continuous_columns])
X_test_encoded[continuous_columns] = scaler_min_max.fit_transform(X_test_encoded[continuous_columns])
```

```python
# Continuous columns often vary in scale, and normalizing them ensures they are on a similar range, preventing any one feature from dominating due to its large magnitude.
# 1. In datasets where continuous features have different ranges, normalization brings all of them into a common scale (0 to 1). For example: Income could range from $10,000 to $100,000, while Age might range from 18 to 70.  Without normalization, models that rely on distance metrics (e.g., k-NN, SVM, neural networks) could be biased toward features with larger values.
# 2. Many machine learning models perform better when the data is normalized, especially algorithms like: Gradient Descent-based models (e.g., neural networks, logistic regression). Distance-based algorithms (e.g., k-NN, SVM, k-Means).
# 3. Reduces Training Time: Normalized features can reduce the time it takes for models to converge during training because they stabilize the optimization process.
```

```python
##-----------------------------------------------------------------------
##  3. Identify and and remove highly correlated columns
##-----------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
corr_matrix = X_train_encoded.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix.round(2), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
corr_matrix
```

```python
from sklearn.neural_network import MLPClassifier
nn_model = MLPClassifier(random_state=42)
#nn_model = MLPClassifier(random_state=42, verbose=True)
#nn_model = MLPClassifier(random_state=42, verbose=True, max_iter=1500)
nn_model = MLPClassifier(random_state=42, verbose=True, max_iter=1500, hidden_layer_sizes=(6,),early_stopping=True)
```

```python
print("Hyperparameters of the MLPClassifier:\n",nn_model.get_params() )
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
##-------------------------------------------------------
##  Train NN model on training data
##-------------------------------------------------------
nn_model.fit(X_train_encoded, y_train)
```

```python
##----------------------------
##  Apply model on test data
##----------------------------
y_prob=nn_model.predict_proba(X_test_encoded)[:,1] # probabilities
#y_prob holds the probability values for each record in the test data
```

```python
y_prob
```

```python
#Create a dataframe to store the test data's Y value and the predicted probability value
#The original admit column is stored as "actual_Y"
#The predicted prob is stored in y_prob
y_pred_df = pd.DataFrame( { "actual_Y": y_test,
"predicted_prob":y_prob})
print(y_pred_df.head())
```

```python
##---------------------------------
##  Based on default cut-off -0.5
##---------------------------------
y_pred_df['predicted_Y'] = y_pred_df.predicted_prob.map(
lambda x: 1 if x >= 0.5 else 0)
print(y_pred_df.head())
```

```python
from sklearn.metrics import confusion_matrix
# Generating the confusion matrix using the DataFrame columns
cm = confusion_matrix(y_pred_df['actual_Y'],y_pred_df['predicted_Y'])
print(cm)
```

```python
# Create a heatmap for the confusion matrix
plt.figure(figsize=(3,3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, 
            xticklabels=['Predicted 0', 'Predicted 1'], 
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
# Calculate Sensitivity (Recall), Specificity, and Accuracy using scikit-learn functions
from sklearn.metrics import  accuracy_score, recall_score, precision_score
sensitivity = recall_score(y_pred_df['actual_Y'], y_pred_df['predicted_Y'])
accuracy = accuracy_score(y_pred_df['actual_Y'], y_pred_df['predicted_Y'])

# Print metrics
print(f"Sensitivity (Recall): {sensitivity:.2f}")
#print(f"Specificity: {specificity:.2f}")
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("\nClassification Report:")
print(classification_report(y_pred_df['actual_Y'],y_pred_df['predicted_Y']))
```
